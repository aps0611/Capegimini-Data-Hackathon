```python
from geopy.geocoders import Nominatim

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class FarmGeocoder:
    def __init__(self):
        self.geocoder = Nominatim(user_agent="FarmGeocoder")

    def geocode_address(self, address):
        location = self.geocoder.geocode(address)
        return (location.latitude, location.longitude) if location else None

# Get the farm address
farm_addresses = list(df_combined.County.unique())

geocoder = FarmGeocoder()

data = []
for address in farm_addresses:
    coordinates = geocoder.geocode_address(address + ',Texas')
    if coordinates:
        latitude, longitude = coordinates
        print(f"Address: {address}")
        print(f"Latitude: {latitude}")
        print(f"Longitude: {longitude}")
        print("---------")
        data.append({'County': address, 'Latitude': latitude, 'Longitude': longitude})
    else:
        print(f"Failed to geocode address: {address}")
        print("---------")

df_geolocation = pd.DataFrame(data)
```

```python
df_geolocation.head(3)
```

```python
import folium

# Create a Folium map centered on the US
map_us = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

# Iterate over each row in the DataFrame
for index, row in df_geolocation.iterrows():
    address = row['County']
    latitude = row['Latitude']
    longitude = row['Longitude']
    
    # Add a marker for each location
    folium.Marker(location=[latitude, longitude], popup=address).add_to(map_us)

# Display the map
map_us
```

```python
def calculate_application_factor(rainfall):
    if rainfall > 200:
      return 0.55 # this factors are for hectares of land
    elif 150 <= rainfall <= 200:
      return 0.5
    elif 100 <= rainfall < 150:
      return 0.45
    elif 50 <= rainfall < 100:
      return 0.4
    elif 25 <= rainfall < 50:
      return 0.4
    else:
        return 0.35

# Convert 'COTTON - ACRES HARVESTED' column to numeric type
df_combined['COTTON - ACRES HARVESTED'] = pd.to_numeric(df_combined['COTTON - ACRES HARVESTED'], errors='coerce')

# Create a new column 'Application Factor' in the DataFrame
df_combined['Application Factor'] = df_combined['Mean Annual\nPrecipitation (cm.)'].apply(calculate_application_factor)

# Multiply the 'Area' column with the 'Application Factor'
df_combined['Lime-tonnes'] = df_combined['COTTON - ACRES HARVESTED'] * df_combined['Application Factor'] * 0.404686
```

E = ((M x FracLime x P X EF) + (M x (1-FracLime) x P x EF)) x Cg/1000

```python
FracLime = 1
P = 0.9 # PURITY OF LIMESTONE
EF = 0.12 # EMISSION FACTOR FOR LIMESTONE
Cg = 3.67

# Calculate the 'E' column:
df_combined['E-cotton-CO2'] = ((df_combined['Lime-tonnes'] * FracLime * P * EF) + (df_combined['Lime-tonnes'] * (1 - FracLime) * P * EF)) * Cg

# Display the updated DataFrame
df_combined
```

```python
# Conversion factor: 1 Ha = 2.47105 acres
conversion_factor = 2.47105

# Convert urea application rate from kg/Ha to kg/acre
df_combined['Urea_tonne_cotton'] = (df_combined['COTTON - ACRES HARVESTED'] * 270 * 0.001) / conversion_factor 

# Display the DataFrame
df_combined
```

```python
EF = 0.20
C = 3.67

df_combined['E-cotton-CO2-urea'] = df_combined['Urea_tonne_cotton'] * EF * C
df_combined
```
